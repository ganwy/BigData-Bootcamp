```python
import pandas as pd
import numpy as np
import json

datos = pd.read_csv("CPdescarga.txt",
                    sep="|",
                    index_col=False,
                    header=1,
                    encoding = "iso-8859-1",
                    dtype=str
                    )
datos.head()
```

```python
from pymongo import MongoClient
con = MongoClient()
db = con.mexico_cod
```

```python
for idx, row in datos[:100000].iterrows():
    db.codigos.insert(json.loads(row.str.encode("utf-8").to_json()))
```

```python
cp = db.codigos.find()
```

```python
for nom in cp:
    print nom["d_asenta"]
```

```python
res = db.codigos.aggregate([{
                        "$group":{"_id":"$d_estado",
                                  "cantidad":{"$sum":1}}
                    }])
for doc in res["result"]:
    print doc
```

```python
from bson.code import Code

map = Code("""function(){
                   emit(this.d_estado, 1);
              }""")

reduce = Code("""function(key, values) { 
                 return Array.sum(values); 
              }""")
mr = db.codigos.map_reduce(map,reduce,"cod_estado", full_response=True)
```
